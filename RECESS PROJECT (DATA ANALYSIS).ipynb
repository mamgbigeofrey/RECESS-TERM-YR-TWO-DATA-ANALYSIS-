```python
#showing dataset correlations using heatmaps

plt.figure(figsize=(10,10))
se.heatmap(dataset.corr(),annot=True,linewidths=0.05,fmt=' .2f',cmap='magma')
plt.show()
```

```python
dataset.plot(figsize=(30,20))
```

```python
#displaying a bar graph of the number of confirmed cases in the world against each country
dataset.plot(kind='bar',x='Country/Region', y='Confirmed', figsize=(130,70))
plt.title('A bar graph showing the number of confirmed cases in the world against each country', color='seagreen',fontsize=100)
plt.show()
```

```python
#visualising the number of confirmed cases for Africa and each country using bar and line graphs


new_df=dataset[dataset['WHO Region']=='Africa']
new_df.plot(kind='bar',x='Country/Region',y='Confirmed',figsize=(20,10),color='brown')
plt.title('A bar graph of the number of COVID_19 Confirmed cases in AFRICA per country',color='seagreen')
plt.show()
# new_df['Total']=new_df.loc[: , ['Confirmed', 'New cases']].sum(axis=1)
new_df.plot(kind='line',x='Country/Region',y='Confirmed',figsize=(20,8))
plt.show()
```

```python
#visualising the number of confirmed cases for South-East Asia and each country using bar and line graphs

new_df=dataset[dataset['WHO Region']=='South-East Asia']
new_df['new']=(new_df['Confirmed']+new_df['New cases'])

new_df.plot(kind='bar',x='Country/Region',y='new',figsize=(20,10),color='gray')
plt.title('A bar graph of the number of COVID_19 Confirmed cases in South-East Asia per country', color='Green')
plt.show()

new_df['new']=(new_df['Confirmed']+new_df['New cases'])
new_df.plot(kind='line',x='Country/Region',y='Confirmed')
plt.show()
```

```python
#performing a pairplot for all the numeric columns in the dataset

se.pairplot(dataset)
plt.show()
```

```python
#getting the highest and lowest affected countires

#getting the country with the list number of covid confirmed cases
data=(dataset['Confirmed']+dataset['New cases']).min()
print('The country with the least number of confirmed cases is ',data,'\n')

#getting the highest number of confirmed cases
data=(dataset['Confirmed']+dataset['New cases']).max()
print('The country with the highest number of confirmed cases is ',data,'\n')

#getting the least number of deaths
data=(dataset['Deaths']+dataset['New deaths']).min()
print('The country with the least number of death cases is ',data,'\n')

#getting the highest number of deaths
data=(dataset['Deaths']+dataset['New deaths']).max()
print('The country with the highest number of death cases is ',data,'\n')

#getting the lowest number of active cases
data=dataset['Active'].min()
print('The country with the lowest number of active cases is ',data,'\n')

#getting the highest number of active cases
data=dataset['Active'].max()
print('The country with the highest number of active cases is ',data,'\n')


```
