```python
#data visualization with pandas
import matplotlib as py
dataset['Deaths'].plot(x='Country/Region',y='Deaths')
dataset.plot()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

new_df=dataset[dataset['WHO Region']=='Africa']
# print(new_df.count())
    
new_df.plot(kind='bar',x='Country/Region',y='Recovered')
plt.title('A bar graph of the number of COVID_19 recovered cases in AFRICA per country')
plt.show()

new_df=dataset[dataset['WHO Region']=='Eastern Mediterranean']
new_df.plot(kind='bar',x='Country/Region',y='Recovered')
plt.show()

new_df=dataset[dataset['WHO Region']=='Europe']
new_df.plot(kind='bar',x='Country/Region',y='Recovered')
plt.show()
```

```python
#showing dataset correlations using heatmaps
import numpy as na
import seaborn as se
plt.figure(figsize=(10,10))
se.heatmap(dataset.corr(),annot=True,linewidths=0.05,fmt=' .2f',cmap='magma')
plt.show()
```

```python
colors = ['lime','blue','maroon','violet','red','yellow','green','white','black','pink','indigo','seagreen','gray','grey','lightgray','lightgrey']
explode = [0.01,0.01,0.01,0.01,0.01]
plt.figure(figsize=(70,70))
plt.pie(new_df['Deaths'].value_counts().values,labels=new_df['Deaths'].value_counts().index,colors=colors,autopct='%.2f')

plt.title('PIE CHART VISUALIZATION OF COVID_19 DEATHS IN EACH COUNTRY',color='blue',fontsize=50)
plt.show()
```
